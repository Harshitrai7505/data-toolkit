# Q3- Explain the following termsgM The difference in npYarray, npYasarray and npYasanyarrayX M The difference between Deep copy and shallow copyX

Sure! Let’s break down each of these concepts.

### 1. Differences between `np.array`, `np.asarray`, and `np.asarray`

- **`np.array`**:
  - **Purpose**: Creates a new NumPy array.
  - **Behavior**: Always copies the input data. It takes an input (like a list or another array) and returns a new array.
  - **Example**:
    ```python
    import numpy as np
    a = np.array([1, 2, 3])
    ```
  
- **`np.asarray`**:
  - **Purpose**: Converts input data into a NumPy array, but does not copy if the input is already a NumPy array.
  - **Behavior**: Returns an array if the input is already an array; otherwise, it creates a new array. It tries to avoid unnecessary copying.
  - **Example**:
    ```python
    b = np.asarray(a)  # No copy, as 'a' is already a NumPy array
    c = np.asarray([1, 2, 3])  # Creates a new array
    ```
  
- **`np.asanyarray`**:
  - **Purpose**: Similar to `np.asarray`, but it allows for subclasses of arrays (like masked arrays or other array-like objects).
  - **Behavior**: If the input is an array subclass (e.g., a masked array), it will return it as is. If it's not, it will convert it to a NumPy array.
  - **Example**:
    ```python
    d = np.asanyarray(np.ma.array([1, 2, 3]))  # Returns a masked array
    ```

### Summary of Differences:
- **Copying**: `np.array` always copies, `np.asarray` avoids copying if possible, and `np.asanyarray` allows subclasses.
- **Return Type**: `np.array` always returns a NumPy array, while `np.asanyarray` can return different types based on the input.

---

### 2. Difference Between Deep Copy and Shallow Copy

- **Shallow Copy**:
  - **Definition**: Creates a new object that is a copy of the original object, but it does not recursively copy objects found within the original object. Instead, it copies references to those objects.
  - **Example**:
    ```python
    import copy

    original = [[1, 2, 3], [4, 5, 6]]
    shallow_copied = copy.copy(original)

    # Modifying the original list
    original[0][0] = 99
    print(shallow_copied)  # Output: [[99, 2, 3], [4, 5, 6]]
    ```
  - **Behavior**: Changes to nested objects in the original will affect the shallow copy because both refer to the same nested objects.

- **Deep Copy**:
  - **Definition**: Creates a new object that is a copy of the original object and recursively copies all objects found within it. This means the new object is entirely independent of the original.
  - **Example**:
    ```python
    import copy

    original = [[1, 2, 3], [4, 5, 6]]
    deep_copied = copy.deepcopy(original)

    # Modifying the original list
    original[0][0] = 99
    print(deep_copied)  # Output: [[1, 2, 3], [4, 5, 6]]
    ```
  - **Behavior**: Changes to nested objects in the original will not affect the deep copy because they are completely independent.

### Summary of Differences:
- **Shallow Copy**: Copies the outer object but shares references to inner objects; changes to inner objects are reflected in both.
- **Deep Copy**: Copies the entire object structure, creating new instances for inner objects; changes do not affect the original.

# Q17- Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

To create a scatter plot using Seaborn that colors points based on their position relative to the origin (quadrants), you can follow the code snippet below:

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(0)

# Create two random arrays
x = np.random.randn(100)
y = np.random.randn(100)

# Determine the quadrant for each point
def determine_quadrant(x, y):
    if x > 0 and y > 0:
        return 'Quadrant I'
    elif x < 0 and y > 0:
        return 'Quadrant II'
    elif x < 0 and y < 0:
        return 'Quadrant III'
    elif x > 0 and y < 0:
        return 'Quadrant IV'
    else:
        return 'Origin'

# Create a DataFrame
data = pd.DataFrame({'x': x, 'y': y})
data['Quadrant'] = data.apply(lambda row: determine_quadrant(row['x'], row['y']), axis=1)

# Create the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='x', y='y', hue='Quadrant', palette='viridis', s=100)

# Add legend, labels, and title
plt.legend(title='Quadrants')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')
plt.axhline(0, color='grey', lw=1)
plt.axvline(0, color='grey', lw=1)

# Show the plot
plt.show()
```

### Explanation:
1. **Libraries**: The code imports `numpy`, `pandas`, `seaborn`, and `matplotlib`.
2. **Random Data**: It generates two arrays of random numbers for the x and y coordinates.
3. **Quadrant Function**: A function determines the quadrant for each point based on its x and y values.
4. **DataFrame**: A Pandas DataFrame is created to hold the x, y, and quadrant data.
5. **Scatter Plot**: `seaborn.scatterplot` is used to create the scatter plot, with points colored based on their quadrant.
6. **Customization**: Axes are labeled, a title is set, and horizontal/vertical lines are added to indicate the axes.

To generate a bar chart of randomly generated categorical data with color-coded bars based on their values and hover tooltips using `Bokeh`, you can use the following code:

```python
from bokeh.io import show
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource, HoverTool
import numpy as np
import pandas as pd

# Generate random categorical data
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(1, 20, size=len(categories))

# Create a DataFrame
data = pd.DataFrame({'Category': categories, 'Value': values})

# Create a ColumnDataSource
source = ColumnDataSource(data=data)

# Create a color mapping based on values
colors = ['#%02x%02x%02x' % (int(255 - (value * 12.75)), int(value * 12.75), 0) for value in values]

# Create a bar chart
p = figure(x_range=categories, height=400, title='Random Categorical Bar Chart', toolbar_location=None)
p.vbar(x='Category', top='Value', source=source, width=0.9, color=colors)

# Add hover tooltips
hover_tool = HoverTool()
hover_tool.tooltips = [("Category", "@Category"), ("Value", "@Value")]
p.add_tools(hover_tool)

# Label the axes
p.xaxis.axis_label = 'Categories'
p.yaxis.axis_label = 'Values'

# Show the plot
show(p)
```
